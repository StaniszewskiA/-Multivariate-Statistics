```r
sales_data <- read.csv("train.csv")
head(sales_data)
```

```r
names(sales_data)
```

```r
dim(sales_data)
```

```r
fit_simple <- lm(sales_data$SalePrice ~ sales_data$LotFrontage)
fit_simple <- lm(SalePrice ~ LotFrontage, data = sales_data)
```

```r
class(fit_simple)
is.list(fit_simple)
names(fit_simple)
```

```r
fit_simple$coefficients
coef(fit_simple)
```

```r
?summary.lm
summary(fit_simple)
```

```r
summaryList <- summary(fit_simple)
summaryList$sigma
summaryList$r.squared
summaryList$fstatistic
```

```r
confint(fit_simple)
```

```r
fit_la <- lm(sales_data$SalePrice ~ sales_data$LotFrontage + sales_data$LotArea)
summary(fit_la)
```

```r
numerical_data <- sales_data[, sapply(sales_data, is.numeric)]

fit_all <- lm(SalePrice ~ ., data = numerical_data)
summary(fit_all)
```

```r
fit_no_yr_old <- update(fit_all, ~ . - YrSold)
summary(fit_no_yr_old)
```

```r

plot(ellipse(fit_la, which = -1), type = "l")
la_coefs <- coef(fit_la)
points(la_coefs[2], la_coefs[3])
```

```r
summary(lm(sales_data$SalePrice ~ sales_data$LotFrontage * sales_data$LotArea))
```

```r
fit_l2 <- lm(sales_data$SalePrice ~ sales_data$LotArea + I(sales_data$LotArea^2))
summary(fit_l2)
```

```r
anova(fit_simple, fit_l2)
```

```r
fit_l5 <- lm(sales_data$SalePrice ~ poly(sales_data$LotArea, 5))
summary(fit_l5)
```
